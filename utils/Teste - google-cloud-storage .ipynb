```python
print("=== CRIANDO CSV LOCAL DE TESTE ===")

csv_local = "teste_local.csv"

with open(csv_local, "w", encoding="utf-8") as f:
    f.write("ANO;EMPRESA;FAZENDA;PRECIP\n")
    f.write("2025;Avant;Fazenda A;120.5\n")
    f.write("2025;Avant;Fazenda B;98.2\n")
    f.write("2026;Avant;Fazenda C;135.7\n")

print(f"Arquivo criado: {os.path.abspath(csv_local)}")
```

```python
print("=== TESTE 1: ARQUIVO LOCAL ===")

try:
    df_local = load_csv_universal("teste_local.csv")
    if df_local is not None:
        print("Sucesso no arquivo local")
        print("Shape:", df_local.shape)
        display(df_local.head())
    else:
        print("Falha ao carregar arquivo local")
except Exception as e:
    print("Erro inesperado no teste local:")
    traceback.print_exc()
```

```python
print("=== TESTE 2: URL HTTP PÚBLICA ===")

url_publica = "https://raw.githubusercontent.com/datasets/gdp/main/data/gdp.csv"

try:
    df_http = load_csv_universal(url_publica)
    if df_http is not None:
        print("Sucesso na URL pública")
        print("Shape:", df_http.shape)
        display(df_http.head())
    else:
        print("Falha ao carregar URL pública")
except Exception as e:
    print("Erro inesperado no teste HTTP:")
    traceback.print_exc()
```

```python
print("=== TESTE 3: URL HTTPS DO GCS ===")

url_gcs_http = "https://storage.googleapis.com/climabd/resumo_2025.csv"

try:
    df_gcs_http = load_csv_universal(url_gcs_http)
    if df_gcs_http is not None:
        print("Sucesso na URL HTTPS do GCS")
        print("Shape:", df_gcs_http.shape)
        display(df_gcs_http.head())
    else:
        print("Falha ao carregar URL HTTPS do GCS")
        print("Isso pode indicar 403, arquivo privado ou URL inválida.")
except Exception as e:
    print("Erro inesperado no teste HTTPS GCS:")
    traceback.print_exc()
```

```python
print("=== TESTE 4: CAMINHO gs:// ===")

gcs_path = "gs://climabd/resumo_2025.csv"

try:
    df_gcs = load_csv_universal(gcs_path)
    if df_gcs is not None:
        print("Sucesso no acesso gs://")
        print("Shape:", df_gcs.shape)
        display(df_gcs.head())
    else:
        print("Falha ao carregar gs://")
        print("Se apareceu erro de credenciais, execute:")
        print("gcloud auth application-default login")
except Exception as e:
    print("Erro inesperado no teste gs://:")
    traceback.print_exc()
```
